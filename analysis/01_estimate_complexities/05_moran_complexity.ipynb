```python
# This is experimental code, that is why it didn't make it to the ecomplexity library yet. Still working on it...
import sys
import numpy as np
import pandas as pd
import networkx as nx
from collections import defaultdict
sys.path.append("../..")
from libraries import network_distance as nd
```

```python
# Loading the input. We need the Product Spaces (both RCA and RPOP)
# We also need MCP exp
ps_rca = nx.read_edgelist("ps_nc_bb_0.01_rca.csv", delimiter = "\t", data = [("count", int), ("significance", float)], nodetype = str)
ps_rpop = nx.read_edgelist("ps_nc_bb_0.1_rpop.csv", delimiter = "\t", data = [("count", int), ("significance", float)], nodetype = str)
mcp_exp = pd.read_csv("../../data/processed/mcp_exp.csv", sep = "\t", dtype = {"prod": str})
mcp_exp
```

```python
# Let's set up some mock parameters that will make it easier to integrate this code into the library
variant = "rpop"
threshold = 1
```

```python
# Since the Product Space is constant, we can cache Q for the calculation of the network variance
# As we'll have to use it for every exporter
mcp_rca = pd.pivot_table(data = mcp_exp, index = "prod", columns = "exporter", values = variant).fillna(0.0) >= threshold

if variant == "rca":
    G = ps_rca
    Q = nd.calculate_spl(ps_rca, ps_rca.nodes, 16)
elif variant == "rpop":
    G = ps_rpop
    Q = nd.calculate_spl(ps_rpop, ps_rpop.nodes, 16)

Q
```

```python
# We now build a dataframe exporter by exporter, by calculating its Moran's I over the Product Space
exp_variance = []
for exporter in mcp_rca:
    # We need the vector to be a probabiity distribution
    exporter_p = mcp_rca[exporter] / mcp_rca[exporter].sum()
    exp_variance.append((
        exporter,
        nd.moran(exporter_p, G, shortest_path_lengths = Q, kernel = "linear")
    ))

exp_variance = pd.DataFrame(data = exp_variance, columns = ("exporter", f"moran_{variant}"))
exp_variance
```

```python
exp_variance.to_csv(f"moran_complexity_{variant}.csv", index = False, sep = "\t")
```
